```python
from src.speech import SpeechInformationRetreiver


sir = SpeechInformationRetreiver(
    model_name='stt_en_fastconformer_hybrid_large_streaming_1040ms',
    lookahead_size=1040,
    decoder_type='rnnt',
    num_keywords=10,
    device='cuda'
)
```

```python
speech_path = 'assets/Folly Audio/Meditation VO - Mindful Beathing.mp3'
speech_segments = sir(speech_path, add_emotion_info=True, max_dist=5., extract_kw=True, verbose=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_segments(segments, add_text=True):
    segments.sort(key=lambda x: x['start'])
    fig, ax = plt.subplots(1, 1, figsize=(15, 5), sharex=True)

    colors = plt.cm.rainbow(np.linspace(0, 1, len(segments)))
    for idx, seg in enumerate(segments):
        rect = plt.Rectangle(
            xy=(seg['start'], 0), 
            width=seg['end'] - seg['start'], 
            height=1, facecolor=colors[idx], alpha=0.5, edgecolor='black'
        )
        ax.add_patch(rect)
        
        if add_text:
            ax.text(
                x=(seg['start'] + seg['end']) / 2, 
                y=0.5, 
                s=seg['text'].replace(', ', '\n'), 
                ha='center', va='center', fontsize=8, fontweight='bold', wrap=True
            )

    ax.set_ylim(0, 1)
    ax.set_yticks([])
    ax.set_ylabel('Segments')
    ax.set_xlim(0, 908)

    plt.tight_layout()
    plt.show()
```

```python
plot_segments(speech_segments, add_text=True)
```
